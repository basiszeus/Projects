```python
import time
import matplotlib.pyplot as plt
```

### 1 - O(1) "Constant Time"

```python
def constant_time_example(arr):
    return arr[0]

array = [1, 2, 3, 4, 5]
print(constant_time_example(array))
```

```python
def constant_time_example(arr):
    return arr[0]

input_sizes = [10, 100, 1000, 10000, 100000, 1000000]
times = []

for size in input_sizes:
    arr = list(range(size))
    start_time = time.time()
    constant_time_example(arr)
    end_time = time.time()
    times.append(end_time - start_time)

plt.plot(input_sizes, times, marker='o')
plt.title('O(1) Complexity')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.ylim(0, max(times) * 1.5)
plt.show()
```

### 2 - O(n) "Linear Time"

```python
def linear_time_example(arr):
    total = 0
    for num in arr:
        total += num
    return total

array = [1, 2, 3, 4, 5]
print(linear_time_example(array)) 
```

### 3 - O(log n) "Logarithmic Time"

```python
def logarithmic_time_example(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

sorted_array = [1, 3, 5, 7, 9, 11, 13, 15]
print(logarithmic_time_example(sorted_array, 7)) 
```

### 4 - O(n²) "Quadratic Time"

```python
def quadratic_time_example(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n):
            print(f"({arr[i]}, {arr[j]})")

array = [1, 2, 3]
quadratic_time_example(array)
```

```python
def quadratic_time_example(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n):
            _ = arr[i] + arr[j]

input_sizes = [10, 100, 500, 1000, 1500, 2000]
times = []

for size in input_sizes:
    arr = list(range(size))
    start_time = time.time()
    quadratic_time_example(arr)
    end_time = time.time()
    times.append(end_time - start_time)

plt.plot(input_sizes, times, marker='o')
plt.title('O(n^2) Complexity')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.ylim(0, max(times) * 1.5)
plt.show()
```

### 5 — O(2^n) "Exponential Time"

```python
def exponential_time_example(n):
    if n <= 1:
        return n
    return exponential_time_example(n-1) + exponential_time_example(n-2)

print(exponential_time_example(10)) 
```

```python
def exponential_time_example(n):
    if n <= 1:
        return n
    return exponential_time_example(n-1) + exponential_time_example(n-2)

input_sizes = range(10, 31, 2)
times = []

for size in input_sizes:
    start_time = time.time()
    exponential_time_example(size)
    end_time = time.time()
    times.append(end_time - start_time)

plt.plot(input_sizes, times, marker='o')
plt.title('O(2^n) Complexity')
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.ylim(0, max(times) * 1.5)
plt.show()
```
